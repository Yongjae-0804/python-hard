### 아시아 국가중 한국, 북한, 일본, 중국의 기대수명 비교하기

#### 1.  df 에서  'Asia'  지역만 추출하여  df_Asia 로 저장하시오.

```python
df_asia = df[df['지역']=='Asia']
df_asia.head()
```

```python
#Asia 대륙의 고유한 국가명을 출력하시오.
df_asia['국가'].unique()
```

```python
# 국가, 연도 별로 그룹화하여  기대수명과 GDP 를 정리한다.
df_asia.groupby(['국가','연도'])[['기대수명','GDP(1인,$)']].mean()
```

```python
# 국가, 연도별 기대수명 피봇테이블을 작성하여  df_life 로 저장한다.
df_life = df_asia.pivot_table(index= '연도',
                          columns = '국가',
                          values = '기대수명'
                         )
df_life.head()
```

```python
df_life.columns
```

```python
# 'South Korea', 'North Korea', 'Japan', 'China' 의 연도별 기대수명을 꺽은선 그래프로 표시하시오
#  범례표시 'Korea, Dem. Rep.', 'Korea, Rep.'

import matplotlib.pyplot as plt
x=df_life.index
plt.figure(figsize=(15,7))
plt.plot(x, df_life['Korea, Dem. Rep.'], label = 'North Korea')
plt.plot(x, df_life['Korea, Rep.'], label = 'South Korea')
plt.plot(x, df_life['Japan'], label = 'Japan')
plt.plot(x, df_life['China'], label = 'China')

plt.title('국가별 기대 수명')
plt.legend()
plt.show()
```

```python
# 국가, 연도별 GDP 피봇테이블을 작성하여  df_gdp 로 저장한다.
df_gdp = df_asia.pivot_table(index= '연도',
                          columns = '국가',
                          values = 'GDP(1인,$)'
                         )
print(df_gdp.head())
```

```python
# 'South Korea', 'North Korea', 'Japan', 'China' 의 연도별 GDP를 꺽은선 그래프로 표시하시오
#  범례표시

import matplotlib.pyplot as plt


x=df_gdp.index
plt.figure(figsize=(15,7))
plt.plot(x, df_gdp['Korea, Dem. Rep.'], label = 'North Korea', color = 'r')
plt.plot(x, df_gdp['Korea, Rep.'], label = 'South Korea', color = 'b')
plt.plot(x, df_gdp['Japan'], label = 'Japan', color = 'pink')
plt.plot(x, df_gdp['China'], label = 'China', color = 'black')
plt.title('일인당 국민 소득($)')

plt.legend()
plt.show()
```
